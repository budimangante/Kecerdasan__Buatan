```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('spotify_churn_dataset.csv')
df.head()
```

```python
#1.Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.
numeric_df = df.select_dtypes(include=['int64', 'float64'])
desc = numeric_df.describe().T
desc[['mean', 'std', 'min', '25%', '50%', '75%', 'max']]
```

```python
#2. 10 record pertama dengan attribut angka
numeric_df.head(10)
```

```python
#3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.
jumlah_record = df.shape[0]

jumlah_attribute = df.shape[1]

nama_attribute = df.columns.tolist()

jumlah_tipe_numeric = numeric_df.shape[1]

print(f"Jumlah record: {jumlah_record}")
print(f"Jumlah attribute: {jumlah_attribute}")
print(f"Jumlah attribute: {nama_attribute}")
print(f"Jumlah attribute: {jumlah_tipe_numeric}")
```

```python
#4. Tampilkan jumlah label dari semua attribute yang bernilai object.
object_df = df.select_dtypes(include=['object'])
for column in object_df.columns:
    print(f"Jumlah label untuk kolom {column}: {object_df[column].nunique()}")
```

```python
#5. Visualisasikan dataset dengan salah satu grafik. Pilih grafik yang paling cocok.
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
df['is_churned'].hist(bins=30, edgecolor='black')
plt.title("is_churned")
plt.xlabel("is_churned")
plt.ylabel("Jumlah Pelanggan")
plt.show()
```

```python
#6. Tampilkan korelasi dataset menggunakan heatmap.
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap Korelasi")
plt.show()
```
